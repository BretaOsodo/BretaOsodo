## Creating Training, Validation and test sets

```python
from sklearn.model_selection import train_test_split

train_val_df, test_df=train_test_split(cancer, test_size=0.2, random_state=42)
train_df, val_df=train_test_split(train_val_df, test_size=0.25,random_state=42)
```

```python
train_df
```

```python
test_df
```

```python
val_df
```

##Identifying inputs and targets for our model

```python
target='diagnosis'
inputs=[col for col in train_df.columns if col not in ['id','diagnosis','Unnamed: 32']]
print (inputs)
```

```python
#creating inputs and targets for the training, validation and test sets
train_inputs= train_df[inputs].copy()
train_target= train_df[target].copy()

val_inputs= val_df[inputs].copy()
val_target= val_df[target].copy()

test_inputs= test_df[inputs].copy()
test_target= test_df[target].copy()
```

##Identifying Numerical columns

```python
## Identifying Numerical columns and Categorical columns

categorical_cols= train_inputs.select_dtypes('object').columns.tolist()
categorical_cols

numerical_cols=[col for col in train_inputs.columns if col not in categorical_cols]
```

```python
numerical_cols
```

##Imputing the data for missing Values

```python
#lets check if we have any missing data
train_inputs[numerical_cols].isna().sum()
```

We don't have any missing missing fdata from our dataset

##Scaling the Numerical Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()

scaler.fit(train_inputs[numerical_cols])
```

```python
# we transform the data
train_inputs[numerical_cols]=scaler.transform(train_inputs[numerical_cols])
val_inputs[numerical_cols]=scaler.transform(val_inputs[numerical_cols])
test_inputs[numerical_cols]=scaler.transform(test_inputs[numerical_cols])
```

## Encoding the categorical column

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False,handle_unknown='ignore')

#we fit the data
encoder.fit(train_inputs[categorical_cols])

#we can generate the columns name for each individual
encoded_cols= list(encoder.get_feature_names_out(categorical_cols))
encoded_cols

#we can transform the encoded cols
train_inputs[encoded_cols]=encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols]=encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols]=encoder.transform(test_inputs[categorical_cols])
```

## save the processed data to disk

```python
#saving the processed data to disk
print(f'train_inputs shape: {train_inputs.shape}')
print(f'train_target shape: {train_target.shape}')
print(f'  val_inputs shape: {val_inputs.shape}')
print(f'  val_target shape: {val_target.shape}')
print(f'  test_inputs shape: {test_inputs.shape}')
print(f'  test_target shape: {test_target.shape}')
```

```python
train_inputs.to_parquet('Train_inputs.parquet')
train_target.to_frame().to_parquet('Train_target.parquet')
val_inputs.to_parquet('Val_inputs.parquet')
val_target.to_frame().to_parquet('Val_target.parquet')
test_inputs.to_parquet('Test_inputs.parquet')
test_target.to_frame().to_parquet('Test_target.parquet')
```

```python
#we can read the dat back using pd.read.parquet
%%time
tain_inputs=pd.read_parquet('Train_inputs.parquet')
train_target=pd.read_parquet('Train_target.parquet')

val_inputs=pd.read_parquet('Val_inputs.parquet')
val_target=pd.read_parquet('Val_target.parquet')

test_inputs=pd.read_parquet('Test_inputs.parquet')
test_target=pd.read_parquet('Test_target.parquet')
```

#Modeling Training an Evaluation

```python
## select the column to be used for training and prediction
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x_train=train_inputs[numerical_cols+encoded_cols]
x_val =val_inputs[numerical_cols + encoded_cols]
x_test=test_inputs[numerical_cols+encoded_cols]
```

##Create and train the model

```python
model= LogisticRegression(solver='liblinear')

#we fit the model
model.fit(x_train,train_target)
```

##Generate Prediction and probabilities

```python
train_preds=model.predict(x_train)
train_probs = model.predict_proba(x_train)
accuracy_score(train_target,train_preds)
```

## a Helper function to predict, compute accuracy and plot confusion matrix

```python
from sklearn.metrics import confusion_matrix

def predict_and_plot(inputs,targets, name=''):
  preds = model.predict(inputs)
  accuracy=accuracy_score(targets,preds)
  print(f'Accuracy of the model is: {accuracy*100}')

  cf=confusion_matrix(targets,preds,normalize='true')
  plt.figure()
  sns.heatmap(cf,annot=True)
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
  plt.title(f'Confusion Matrix for {name}')
  plt.show()
  return preds
```

##Evaluate on validation and test set

```python
val_preds= predict_and_plot(x_val,val_target,'Validation')
test_preds=predict_and_plot(x_test,test_target,'Test')
```
